```python
name=['Phillippines','Indonesia','Malta','India']
defective=[29,33,31,20]
error_free=[271,267,269,280]
```

```python
data = {'Name': name, 'defective': defective,'error_free':error_free}
df = pd.DataFrame(data)
df
```

```python
defective1=df['defective']
print(defective1.mean())
print(round(defective1.std(),2))
```

```python
error_free1=df['error_free']
print(error_free1.mean())
print(round(error_free1.std(),2))
```

```python
#t_statistic,p_value=stats.ttest.ind
t_statistics,p_values=stats.ttest_ind(defective1,error_free1)
p_values
```

```python
if p_values<0.05:
  print("reject null hypothesis")
else:
  print("fail to reject null hypothesis")
```
